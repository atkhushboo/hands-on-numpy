```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "sales": [100, 120, 90, 150, 180, 200],
    "profit": [20, 25, 15, 35, 40, 50],
    "customers": [50, 55, 45, 70, 80, 90]
})
```

```python
# Q1. Sales ka line chart banao.
plt.plot(df["sales"],color="red",label="sales")
plt.xlabel("Month")
plt.ylabel("sales")
plt.title("Monthly Sales")
plt.legend()
plt.show()
```

```python
# Sales ka bar chart banao.

plt.bar(df["month"],
    df["sales"])
plt.show()
```

```python

#  Profit ka histogram banao.

plt.hist(df["profit"],bins=3,color='pink')
plt.title("Profit chart")
# plt.grid(True)
plt.show()
```

```python
#customer v/s sales

plt.scatter(df["customers"],df['sales'])
plt.grid(True)
plt.show()

```
