```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "sales": [100, 120, 90, 150, 180, 200],
    "profit": [20, 25, 15, 35, 40, 50],
    "customers": [50, 55, 45, 70, 80, 90]
})
```

```python
# Q1. Sales ka line chart banao.
plt.plot(df["sales"],color="red")
plt.xlabel("Month")
plt.ylabel("sales")
plt.title("Monthly Sales")
plt.show()
```

```python
# Sales ka bar chart banao.

plt.bar(df["month"],
    df["sales"])
plt.show()
```
